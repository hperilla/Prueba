```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 3/2, .01)
p1=(x**4)-2*(x**3)
plt.figure(figsize=(10, 5))
plt.plot(x, p1, color='red', linestyle='-',label='x−2x**3',linewidth=2,markersize=5)
legend = plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```
